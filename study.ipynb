# Estudio para pasar pruebas de fondeo

En este estudio veremos la probabilidad de que una estrategia pase una prueba de fondeo en base a los siguientes parámetros:

* **Tasa de aciertos**
* **Promedio de la operación ganadora (en porcentaje)**
* **Promedio de la operación perdedora (en porcentaje)**
* **Balance inicial**
* **Objetivo para pasar la prueba de fondeo**
* **Objetivo para perder la prueba de fondeo**

**IMPORTANTE**: Se asume que la pérdida máxima diaria no se alcanza. Debes ser tú quien ajuste el riesgo en base a las operaciones por día.

```python
tasa_aciertos = 46
promedio_operacion_ganadora = 2
promedio_operacion_perdedora = 1
balance_inicial = 200000
objetivo = 220000
objetivo_perdida = 180000

import utils
```

```python
results, successes, failures = utils.run_simulations(
    1000, 
    tasa_aciertos, 
    promedio_operacion_ganadora, 
    promedio_operacion_perdedora, 
    balance_inicial, 
    objetivo, 
    objetivo_perdida
)
```

### Apariencia pruebas exitosas

```python
utils.draw_results(successes[0:7], balance_inicial)
```

### Apariencia pruebas fallidas

```python
utils.draw_results(failures[0:5], balance_inicial)
```

### Prueba de fondeo más corta

```python
shortest = utils.get_shortest(results, balance_inicial)
num_operations_shortest = len(shortest)
print(f"La prueba de fondeo más corta (exitosa) se ha hecho en {num_operations_shortest} operaciones")
```

```python
utils.draw_results([shortest], balance_inicial)
```

### Prueba de fondeo más larga (exitosa)

```python
longest = utils.get_longest(results, balance_inicial)
num_operations_longest = len(longest)
print(f"La prueba de fondeo más larga (exitosa) se ha hecho en {num_operations_longest} operaciones")
```

```python
utils.draw_results([longest], balance_inicial)
```

### Duración media de las pruebas (exitosas)

```python
average_duration_success = utils.get_average_duration(successes)
print(f"La prueba de fondeo (exitosa) dura de media {average_duration_success} operaciones")
```

### Duración media de las pruebas (fallidas)

```python
average_duration_failed = utils.get_average_duration(failures)
print(f"La prueba de fondeo (fallida) dura de media {average_duration_failed} operaciones")
```
